```python
# checking null

def nan_check(df):
    return round((100*df.isnull().sum())/len(df),2).sort_values(ascending=False)
```

```python
nan_df=nan_check(churn)
nan_df
```

```python
#dropping unnecessary columns

unnecessary=["date_of_last_rech_data_6","date_of_last_rech_data_7","date_of_last_rech_data_9","date_of_last_rech_data_8","date_of_last_rech_9",
             "date_of_last_rech_8","date_of_last_rech_7","last_date_of_month_9","date_of_last_rech_6","last_date_of_month_8",
            "last_date_of_month_7","last_date_of_month_6","circle_id","mobile_number"]
len(unnecessary)
```

```python
churn2=churn.drop(unnecessary,axis=1)
```

```python
churn2.shape
```

```python
rech_col = [col for col in churn2.columns if "rech" in col or "fb" in col or "night" in col or "sacget" in col]
rech_col
```

```python
churn2[rech_col]=churn[rech_col].fillna(0)

```

```python
# again check nulls

nan_df=nan_check(churn2)
```

```python
# extracting column which have null perc greater than 30

col_grt_30=nan_df[nan_df>30].index
col_grt_30
```

```python
churn2.drop(col_grt_30,axis=1,inplace=True)
```

```python
# again check nulls

nan_check(churn2)
```

## extracting High customers

```python
#calculating total recharge amount for data for 6th and 7th month
churn2["total_rech_amnt_data_6"]=churn2["total_rech_data_6"] * churn2["av_rech_amt_data_6"]
churn2["total_rech_amnt_data_7"]=churn2["total_rech_data_7"] * churn2["av_rech_amt_data_7"]
```

```python
#calculating total recharge amount (data+minute) for 6th and 7th month
churn2["Grand_total_rech_6"]=churn2["total_rech_amt_6"] + churn2["total_rech_amnt_data_6"]
churn2["Grand_total_rech_7"]=churn2["total_rech_amt_7"] + churn2["total_rech_amnt_data_7"]
```

```python
#calculating average recharge amount in 6th an 7th month
churn2["avg_rech_amnt_6_7"]=(churn2["Grand_total_rech_7"]+churn2["Grand_total_rech_6"])/2
```

```python
x=churn2["avg_rech_amnt_6_7"].quantile(0.7)
x
```

```python
high_value_df = churn2[churn2["avg_rech_amnt_6_7"]>=x]
high_value_df
```

```python
#dropping above drived  columns

high_value_df.drop(["total_rech_amnt_data_6","total_rech_amnt_data_7","Grand_total_rech_6","Grand_total_rech_7","avg_rech_amnt_6_7"],axis=1,inplace=True)
```

```python
high_value_df.shape
```

## Imputation Technique

```python
#import knn imputor

from sklearn.impute import KNNImputer
```

```python
knn=KNNImputer()
```

```python
high_val_knn_df=knn.fit_transform(high_value_df)
```

```python

high_val_knn_df = pd.DataFrame(high_val_knn_df,columns=high_value_df.columns)
```

```python
nan_check(high_val_knn_df)
```

## Tag churners

```python
#condition fo churners i.e if all are zero then he is churner otherwise not

cond= (high_val_knn_df["total_ic_mou_9"]==0) & (high_val_knn_df["total_og_mou_9"]==0) &  (high_val_knn_df["vol_2g_mb_9"]==0) & (high_val_knn_df["vol_3g_mb_9"]==0)
      
```

```python
high_val_knn_df["churners"] = np.where(cond,1,0)
```

```python
#removing all 9 month columns as in case study

month9=[i for i in high_val_knn_df.columns if "9" in i]
month9.append("sep_vbc_3g")
```

```python
high_val_knn_df.drop(month9,axis=1,inplace=True)
```
